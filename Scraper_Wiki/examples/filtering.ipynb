# Filtragem de Artigos

Demonstra como selecionar artigos por categoria ou idioma e exportar subconjuntos do dataset.

```python
import json, pandas as pd

with open('datasets_wikipedia_pro/qa_pairs.json') as f:
    data = json.load(f)

df = pd.DataFrame(data)
```

```python
# Selecionar artigos em portugues sobre Programacao
subset = df[(df['language']=='pt') & (df['category']=='Programacao')]
subset.to_json('programacao_pt.json', orient='records', force_ascii=False)
```

```python
# Filtrar por outra categoria ou idioma
en_math = df[(df['language']=='en') & (df['category']=='Mathematics')]
en_math.to_csv('math_en.csv', index=False)
```
